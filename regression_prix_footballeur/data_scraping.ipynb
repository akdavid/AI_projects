# Récupération des données

```python
headings = ['Rang', 'Joueur', 'Poste', 'Age', 'Nat.', 'Club', 'Valeur', 'Matchs', 'Buts', 'Buts_CSC', 'Passes_d', 
             'Cartons_jaunes', 'Second_avertissement', 'Cartons_rouges', 'Entre', 'Remplace', 'Buts_2021', 'Passes_2021', 'Minutes_2021', 
            'Buts_2022', 'Passes_2022', 'Minutes_2022', 'Buts_total', 'Passes_total', 'Minutes_total']
```

```python
# getting the table body

# list to store all content
content = []

for i in range(1, 21):
    headers = random.choice(headers_list) 
    URL = f'https://www.transfermarkt.us/spieler-statistik/wertvollstespieler/marktwertetop?ajax=yw1&altersklasse=alle&ausrichtung=Sturm&jahrgang=0&kontinent_id=0&land_id=0&page={i}&plus=1&spielerposition_id=alle'

    # get web data
    page = requests.get(URL, headers=headers)
    # parse web data
    soup = BeautifulSoup(page.content, "html.parser")
    # getting the table body
    html_tbody = soup.find_all('tbody')[-1]
    #getting all the rows in table body
    html_text = [tr for tr in html_tbody.find_all('tr')]
    
    # loop through table body
    for tr in html_text:
        # getting all th, td
        th = tr.find_all(['th','td'])
        # storing all th value in row and removing white space
        row = [i.text.strip() for i in th]

        if len(row)==18:
            row.pop(1)
            row.pop(1)
            for year in [2021, 2022, 'ges']:
                str_url = tr.find(class_="rechts hauptlink").find('a')['href']    # get the url of the main page of the player
                URL_stats = 'https://www.transfermarkt.us'+str_url.replace('marktwertverlauf', 'leistungsdaten')+f'/plus/0?saison={str(year)}'  # get the url of the stats page
        
                headers_stats = random.choice(headers_list) 
                page_stats = requests.get(URL_stats, headers=headers_stats)
                soup_stats = BeautifulSoup(page_stats.content, "html.parser")
                html_tfoot = soup_stats.find_all('tfoot')[0]
                stats_data = html_tfoot.find_all(class_=["zentriert", "rechts"])
                stats_data_regex = [re.sub("<(?:\"[^\"]*\"['\"]*|'[^']*'['\"]*|[^'\">])+>", '', str(stats_data[i])) for i in range(len(stats_data))]
                
                buts, passes, minutes = [stats_data_regex[2], stats_data_regex[3], stats_data_regex[7]]
                
                row.append(buts)
                row.append(passes)
                row.append(minutes)

            content.append(row)        
```

```python
# save contents in a dataframe
data = pd.DataFrame(content[:], columns=headings)
```

# Mise en forme et analyse des données

```python
# check few top rows of data
data.head()
```

```python
data = data.drop(columns=['Nat.', 'Club', 'Matchs', 'Buts', 'Buts_CSC', 'Passes_d', 
             'Cartons_jaunes', 'Second_avertissement', 'Cartons_rouges', 'Entre', 'Remplace'])
```

```python
# remove € symbol and m from Valeur and convert to float
# convert Age, Buts, Passes, and Minutes to int
# replace '-' values by 0
# translate the poste in french

data['Valeur'] = data['Valeur'].str.replace('€','')
data['Valeur'] = data['Valeur'].str.replace('m','')
data['Valeur'] = data['Valeur'].astype(float)
data['Age'] = data['Age'].astype(int)


data['Buts_2021'] = data['Buts_2021'].str.replace('-','0')
data['Buts_2022'] = data['Buts_2022'].str.replace('-','0')
data['Buts_total'] = data['Buts_total'].str.replace('-','0')
data['Passes_2021'] = data['Passes_2021'].str.replace('-','0')
data['Passes_2022'] = data['Passes_2022'].str.replace('-','0')
data['Passes_total'] = data['Passes_total'].str.replace('-','0')

data['Buts_2021'] = data['Buts_2021'].astype(int)
data['Buts_2022'] = data['Buts_2022'].astype(int)
data['Buts_total'] = data['Buts_total'].astype(int)

data['Passes_2021'] = data['Passes_2021'].astype(int)
data['Passes_2022'] = data['Passes_2022'].astype(int)
data['Passes_total'] = data['Passes_total'].astype(int)

data['Minutes_2021'] = data['Minutes_2021'].str.replace("'",'')
data['Minutes_2021'] = data['Minutes_2021'].str.replace(".",'')
data['Minutes_2021'] = data['Minutes_2021'].astype(int)

data['Minutes_2022'] = data['Minutes_2022'].str.replace("'",'')
data['Minutes_2022'] = data['Minutes_2022'].str.replace(".",'')
data['Minutes_2022'] = data['Minutes_2022'].astype(int)

data['Minutes_total'] = data['Minutes_total'].str.replace("'",'')
data['Minutes_total'] = data['Minutes_total'].str.replace(".",'')
data['Minutes_total'] = data['Minutes_total'].astype(int)

data['Poste'] = data['Poste'].str.replace('Centre-Forward', 'Avant-Centre')
data['Poste'] = data['Poste'].str.replace('Left Winger', 'Ailier Gauche')
data['Poste'] = data['Poste'].str.replace('Right Winger', 'Ailier Droit')
data['Poste'] = data['Poste'].str.replace('Second Striker', 'Deuxième Attaquant')
```
